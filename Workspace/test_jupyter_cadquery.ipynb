```python
import cadquery as cq

result = (
    cq.Workplane()
    .sketch()
    .trapezoid(4, 3, 90)
    .vertices()
    .circle(0.5, mode="s")
    .reset()
    .vertices()
    .fillet(0.25)
    .reset()
    .rarray(0.6, 1, 5, 1)
    .slot(1.5, 0.4, mode="s", angle=90)
    .finalize()
    .extrude(2)
)
display(result)
```
