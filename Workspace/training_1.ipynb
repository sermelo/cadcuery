```python
import cadquery as cq
```

```python
length = 80.0
height = 60.0
thickness = 10.0
center_hole_dia = max(length, height) / 3

result = (
    cq.Workplane("front").
    box(length, height, thickness).
    faces(">Z").
    hole(center_hole_dia)
)
result
```

```python
result = (
    cq.Workplane("front").
    rect(2, 2).
    circle(5).
    extrude(0.5)
)
result
```

```python
result = (
    cq.Workplane().
    lineTo(2.0, 0).
    lineTo(2.0, 1.0).
    threePointArc((1.0, 1.5), (0.0, 1.0)).
    close().
    extrude(0.2)
)
result
```
